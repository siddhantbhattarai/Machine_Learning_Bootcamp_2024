## **Introduction**

### **About the Dataset**
This dataset contains health-related information for predicting **liver cancer stages**. It includes a mix of **numerical** and **categorical** features, along with some synthetic challenges to simulate real-world data cleaning tasks. The dataset has 15,000 records, of which 3,000 are duplicates, and contains:
- Missing values (both partial and fully null columns).
- Unnecessary columns irrelevant to the task.
- Target labels divided into multiple binary columns (`Stage_I`, `Stage_II`, etc.), which need to be merged into a single `Cancer Stage` column.

### **Features**
1. **Age**: Patient's age (numerical).
2. **Gender**: Patient's gender (categorical: Male/Female).
3. **ALT (U/L)**, **AST (U/L)**: Liver enzymes (numerical).
4. **Total Bilirubin (mg/dL)**, **Direct Bilirubin (mg/dL)**: Indicators of liver function (numerical).
5. **Albumin (g/dL)**: Blood protein level (numerical).
6. **Alcohol Consumption**: Indicates if the patient consumes alcohol (categorical: Yes/No).
7. **Hepatitis B Infection**, **Cirrhosis History**: History of liver-related conditions (categorical: Yes/No).
8. **Unnecessary Columns**: Random data included to simulate irrelevant features.
9. **Cancer Stage**: Target labels divided into multiple columns (`Stage_I`, `Stage_II`, `Stage_III`, `Stage_IV`).

The following tasks will guide us through the **data cleaning process**:

### **1. Dataset Exploration**
- Understand the data structure with `.info()`, `.describe()`, and `.head()`.

### **2. Handling Missing Values**
- Drop **columns with all null values**.
- Drop **rows with critical null values** instead of imputing them.

### **3. Removing Duplicates**
- Identify and remove duplicate rows to avoid biased models.

### **4. Dropping Unnecessary Columns**
- Remove columns that do not contribute to predicting the target (`Unnecessary_Column_1`, `Unnecessary_Column_2`).

### **5. Merging Target Columns**
- Combine the split target columns (`Stage_I`, `Stage_II`, etc.) into a single `Cancer Stage` column.

### **6. Encoding Categorical Variables**
- Use **Label Encoding** for binary columns (e.g., Gender).
- Use **One-Hot Encoding** if required for the target column or multi-class features.

### **7. Scaling Numerical Features**
- Scale numerical columns (e.g., ALT, AST) using **Standard Scaler**, **MinMax Scaler**, or **Robust Scaler** to standardize the dataset for machine learning models.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/siddhantbhattarai/Machine_Learning_Bootcamp_2024/refs/heads/main/Datasets/liver_cancer_dataset.csv')
df.head()
```

```python
# 1. Basic properties
# Shape of the dataset: Number of rows and columns
print("Shape of the dataset:", df.shape)
```

```python
# Data types and null values
print("\nData types and null values:")
print(df.info())
```

```python
# 2. Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# 3. Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# 4. Statistical summary
# Useful for numerical data
print("\nStatistical summary:")
display(df.describe())
```

```python
# 5. Unique values in each column
print("\nUnique values in each column:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()} unique values")
```

```python
# 6. Distribution of target columns
print("\nDistribution of target columns (Stage columns):")
print(df[['Stage_I', 'Stage_II', 'Stage_III', 'Stage_IV']].sum())
```

```python
# Drop duplicates
df = df.drop_duplicates()

# Confirm duplicates are removed
print("Number of duplicate rows after removal:", df.duplicated().sum())
```

```python
# Drop columns with all null values
df = df.dropna(axis=1, how='all')

# Drop rows with any null values (if critical features have nulls)
df = df.dropna(axis=0, how='any')

# Confirm no null values remain
print("Missing values after dropping rows and columns:")
print(df.isnull().sum())
```

```python
# Drop irrelevant columns
columns_to_drop = ["Unnecessary_Column_1", "Unnecessary_Column_2"]
df = df.drop(columns=columns_to_drop, axis=1)

# Confirm the columns are removed
print("Remaining columns:", df.columns)
```

```python
# Merge target columns into a single "Cancer Stage" column
df['Cancer Stage'] = df[['Stage_I', 'Stage_II', 'Stage_III', 'Stage_IV']].idxmax(axis=1)

# Drop original stage columns
df = df.drop(columns=['Stage_I', 'Stage_II', 'Stage_III', 'Stage_IV'])

# Confirm the target column
print(df['Cancer Stage'].head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding for binary categorical columns
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Alcohol Consumption'] = label_encoder.fit_transform(df['Alcohol Consumption'])
df['Hepatitis B Infection'] = label_encoder.fit_transform(df['Hepatitis B Infection'])
df['Cirrhosis History'] = label_encoder.fit_transform(df['Cirrhosis History'])
df['Cancer Stage'] = label_encoder.fit_transform(df['Cancer Stage'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import joblib  # To save the best model
```

```python
# Features and target
X = df.drop(columns=['Cancer Stage'])
y = df['Cancer Stage']
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Logistic Regression Model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_scaled, y_train)
```

```python
# Predictions and Evaluation
y_pred_logreg = logreg.predict(X_test_scaled)
logreg_report = classification_report(y_test, y_pred_logreg, zero_division=0)
logreg_cm = confusion_matrix(y_test, y_pred_logreg)

# Print results
print("Logistic Regression Evaluation:")
print(logreg_report)
print("Confusion Matrix:")
print(logreg_cm)
```

```python
# Random Forest Model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)
```

```python
# Predictions and Evaluation
y_pred_rf = rf.predict(X_test_scaled)
rf_report = classification_report(y_test, y_pred_rf, zero_division=0)
rf_cm = confusion_matrix(y_test, y_pred_rf)

# Print results
print("Random Forest Evaluation:")
print(rf_report)
print("Confusion Matrix:")
print(rf_cm)
```

```python
# K-Nearest Neighbors Model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
```

```python
# Predictions and Evaluation
y_pred_knn = knn.predict(X_test_scaled)
knn_report = classification_report(y_test, y_pred_knn, zero_division=0)
knn_cm = confusion_matrix(y_test, y_pred_knn)

# Print results
print("K-Nearest Neighbors Evaluation:")
print(knn_report)
print("Confusion Matrix:")
print(knn_cm)
```

```python
# Support Vector Machine Model
svm = SVC(random_state=42)
svm.fit(X_train_scaled, y_train)
```

```python
# Predictions and Evaluation
y_pred_svm = svm.predict(X_test_scaled)
svm_report = classification_report(y_test, y_pred_svm, zero_division=0)
svm_cm = confusion_matrix(y_test, y_pred_svm)

# Print results
print("Support Vector Machine Evaluation:")
print(svm_report)
print("Confusion Matrix:")
print(svm_cm)
```

```python
# Gradient Boosting Model
gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train_scaled, y_train)
```

```python
# Predictions and Evaluation
y_pred_gb = gb.predict(X_test_scaled)
gb_report = classification_report(y_test, y_pred_gb, zero_division=0)
gb_cm = confusion_matrix(y_test, y_pred_gb)

# Print results
print("Gradient Boosting Evaluation:")
print(gb_report)
print("Confusion Matrix:")
print(gb_cm)
```

```python
# Store models and their evaluation results
models = {
    "Logistic Regression": logreg_report,
    "Random Forest": rf_report,
    "K-Nearest Neighbors": knn_report,
    "Support Vector Machine": svm_report,
    "Gradient Boosting": gb_report
}
```

```python
# Calculate weighted average F1-scores using zero_division=0 to avoid precision issues
f1_scores = {
    "Logistic Regression": classification_report(y_test, y_pred_logreg, output_dict=True, zero_division=0)['weighted avg']['f1-score'],
    "Random Forest": classification_report(y_test, y_pred_rf, output_dict=True, zero_division=0)['weighted avg']['f1-score'],
    "K-Nearest Neighbors": classification_report(y_test, y_pred_knn, output_dict=True, zero_division=0)['weighted avg']['f1-score'],
    "Support Vector Machine": classification_report(y_test, y_pred_svm, output_dict=True, zero_division=0)['weighted avg']['f1-score'],
    "Gradient Boosting": classification_report(y_test, y_pred_gb, output_dict=True, zero_division=0)['weighted avg']['f1-score']
}

# Print the F1-scores for comparison
for model, f1_score in f1_scores.items():
    print(f"{model} Weighted F1-Score: {f1_score:.4f}")
```

```python
# Plotting the F1-scores of all models
plt.figure(figsize=(10, 6))
plt.bar(f1_scores.keys(), f1_scores.values(), color=['blue', 'green', 'red', 'orange', 'purple'])

# Adding titles and labels
plt.title("Comparison of F1-Scores for Different Models", fontsize=16)
plt.xlabel("Machine Learning Models", fontsize=14)
plt.ylabel("Weighted Average F1-Score", fontsize=14)

# Displaying the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Find the best model based on the highest F1-score
best_model_name = max(f1_scores, key=f1_scores.get)
print(f"Best model based on F1 score: {best_model_name}")
```

```python
# Select the best model based on F1-score
if best_model_name == "Logistic Regression":
    best_model = logreg
elif best_model_name == "Random Forest":
    best_model = rf
elif best_model_name == "K-Nearest Neighbors":
    best_model = knn
elif best_model_name == "Support Vector Machine":
    best_model = svm
else:
    best_model = gb

# Save the best model using joblib
joblib.dump(best_model, 'best_model.pkl')
print(f"Best model {best_model_name} saved as 'best_model.pkl'.")
```

```python
import numpy as np
# Display confusion matrix for the best model
best_cm = confusion_matrix(y_test, best_model.predict(X_test_scaled))
sns.heatmap(best_cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title(f'Confusion Matrix of Best Model: {best_model_name}')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### **Project Summary: Liver Cancer Prediction using Machine Learning**

This project aims to predict the **cancer stage** of patients based on several medical features using various machine learning models. The entire workflow includes the following key stages:

### **1. Data Generation and Preparation:**
- **Dataset Creation**: A synthetic dataset was generated to simulate health data for liver cancer prediction. The dataset contains both categorical and numerical features, including:
  - **Numerical Features**: Age, ALT (U/L), AST (U/L), Total Bilirubin (mg/dL), etc.
  - **Categorical Features**: Gender, Alcohol Consumption, Hepatitis B Infection, Cirrhosis History.
  - **Target Variable**: `Cancer Stage`, which represents the stage of liver cancer (from Stage I to IV).

### **2. Data Preprocessing:**
- **Handling Null Values**: Null values were introduced, and strategies were employed to handle them. Since many of the rows were null, the rows with null target values were dropped.
- **Duplicate Data Removal**: A significant number of duplicate rows (3000+) were added to simulate real-world issues, which were then removed.
- **Feature Encoding**:
  - **Label Encoding**: Used to convert categorical variables (e.g., Gender, Alcohol Consumption) into numeric labels.
  - **One-Hot Encoding**: Applied to categorical columns such as `Cancer Stage` to create binary features for each cancer stage.
- **Scaling**: Data normalization was applied using `StandardScaler` to standardize numerical features, ensuring that models could process them without being biased by the scale of the features.

### **3. Model Selection:**
Five different machine learning algorithms were chosen to predict the liver cancer stage:
- **Logistic Regression**
- **Random Forest**
- **K-Nearest Neighbors (KNN)**
- **Support Vector Machine (SVM)**
- **Gradient Boosting**

### **4. Model Training and Evaluation:**
- **Model Training**: Each model was trained on the preprocessed data, and predictions were made on the test set.
- **Evaluation Metrics**: Multiple evaluation metrics were used to assess model performance:
  - **Accuracy**
  - **Precision**
  - **Recall**
  - **F1-Score** (particularly the weighted average F1-score)
- **Handling Undefined Metrics**: In cases where models did not predict certain classes, we used `zero_division=0` to avoid warnings and ensure a more reliable F1-score calculation.

### **5. Model Comparison:**
- After training and evaluation, the **F1-scores** of all models were compared using a dictionary that stored the weighted average F1-score for each model.
- The model with the highest F1-score was selected as the **best model**.

### **6. Best Model Selection and Saving:**
- The best model (based on the highest F1-score) was saved using `joblib.dump`, making it ready for future use or deployment.
- **Confusion Matrix**: A confusion matrix was displayed for the best model, providing insight into its classification performance. The matrix was visualized using a heatmap for better understanding.

### **7. Final Visualization:**
- **Confusion Matrix Visualization**: After selecting the best model, we visualized its performance using a heatmap of the confusion matrix, which shows how well the model classified the different cancer stages (Stage I to Stage IV).
- **Model Comparison Plot**: A comparison of the models' F1-scores was plotted to visually show which model performed best.

---

### **Key Challenges and Solutions:**
- **Class Imbalance**: Some models had issues with precision for certain classes, especially when the target class wasn’t predicted at all. We handled this by setting `zero_division=0` in evaluation metrics.
- **Data Cleaning**: Handling missing values, duplicates, and irrelevant features was essential to ensure the models learned from the most relevant data.
- **Feature Scaling and Encoding**: Ensuring numerical features were scaled and categorical features were properly encoded helped in making the models work efficiently.
