```python
import pandas as pd

url = "https://raw.githubusercontent.com/siddhantbhattarai/Machine_Learning_Bootcamp_2024/refs/heads/main/Datasets/stroke_prediction_dataset.csv"
df = pd.read_csv(url)
df.head()
```

```python
# 1. Basic properties
# Shape of the dataset: Number of rows and columns
print("Shape of the dataset:", df.shape)
```

```python
# Data types and null values
print("\nData types and null values:")
print(df.info())
```

```python
# 2. Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# 3. Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# 4. Statistical summary
# Useful for numerical data
print("\nStatistical summary:")
display(df.describe())
```

```python
# 5. Unique values in each column
print("\nUnique values in each column:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()} unique values")
```

```python
# Drop columns with all null values
df = df.dropna(axis=1, how='all')

# Drop rows with any null values (if critical features have nulls)
df = df.dropna(axis=0, how='any')

# Confirm no null values remain
print("Missing values after dropping rows and columns:")
print(df.isnull().sum())
```

```python
# Function to extract HDL and LDL values
def extract_cholesterol(cholesterol):
    # Split the string by comma and extract values
    try:
        hdl, ldl = cholesterol.split(', ')
        hdl_value = int(hdl.split(': ')[1])
        ldl_value = int(ldl.split(': ')[1])
        return hdl_value, ldl_value
    except:
        return None, None  # In case of an error (e.g., missing data)
```

```python
# Apply the extraction function to the "Cholesterol Levels" column
df[['HDL', 'LDL']] = df['Cholesterol Levels'].apply(lambda x: pd.Series(extract_cholesterol(x)))
```

```python
# If you want to drop the original "Cholesterol Levels" column after separating:
df.drop('Cholesterol Levels', axis=1, inplace=True)
```

```python
display(df.head())
```

```python
print("# Check if any missing or invalid values exist after extraction:")
print(df.isnull().sum())
```

```python
# Split the Symptoms column by commas and store them as lists
# We first clean up any extra spaces and split by commas
df['Symptoms'] = df['Symptoms'].apply(lambda x: [s.strip() for s in x.split(',')] if isinstance(x, str) else [])
```

```python
# Display the first few rows of the updated DataFrame
df.head()
```

```python
# Use MultiLabelBinarizer to one-hot encode the Symptoms
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
```

```python
# Fit and transform the Symptoms column
symptom_encoded = mlb.fit_transform(df['Symptoms'])
```

```python
# Convert the encoded symptoms into a DataFrame with proper column names
symptom_df = pd.DataFrame(symptom_encoded, columns=mlb.classes_)
```

```python
# Concatenate the original DataFrame with the new one-hot encoded symptom columns
df = pd.concat([df, symptom_df], axis=1)
```

```python
# Drop the original Symptoms column (optional)
df.drop('Symptoms', axis=1, inplace=True)
```

```python
# Display the first few rows of the updated DataFrame
print(df.head())
```

```python
# Drop unnecessary columns
df = df.drop(['Patient ID', 'Patient Name'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Identify the object columns (categorical columns)
object_columns = df.select_dtypes(include=['object']).columns

# Apply Label Encoding to each object column
for column in object_columns:
    df[column] = label_encoder.fit_transform(df[column])

# Verify the changes
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Separate target and features
X = df.drop(columns=['Diagnosis'])  # Features (all columns except 'Diagnosis')
y = df['Diagnosis']  # Target variable (Diagnosis)
```

```python
# 2. Handle missing values
X = X.fillna(X.mean())  # Impute missing values in feature columns with the mean of the column
y = y.fillna(y.mode()[0])  # Impute missing values in target with the mode (most frequent value)
```

```python
# 3. Normalize the data (standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Standardize the feature set
```

```python
# 4. Train-Test Split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Function to train and evaluate models
def train_and_evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Evaluation Metrics
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    
    # Print Results
    print(f"Model: {model.__class__.__name__}")
    print(f"Accuracy: {accuracy:.4f}")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("Classification Report:")
    print(class_report)
    
    # Visualize Confusion Matrix
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['No Stroke', 'Stroke'], yticklabels=['No Stroke', 'Stroke'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix for {model.__class__.__name__}')
    plt.show()
```

```python
# 5. Train and evaluate models

# Logistic Regression
logreg_model = LogisticRegression()
train_and_evaluate_model(logreg_model, X_train, X_test, y_train, y_test)
```

```python
# Decision Tree Classifier
dt_model = DecisionTreeClassifier(random_state=42)
train_and_evaluate_model(dt_model, X_train, X_test, y_train, y_test)
```

```python
# Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
train_and_evaluate_model(rf_model, X_train, X_test, y_train, y_test)
```

```python
# Support Vector Machine (SVM)
svm_model = SVC(random_state=42)
train_and_evaluate_model(svm_model, X_train, X_test, y_train, y_test)
```

```python
# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier()
train_and_evaluate_model(knn_model, X_train, X_test, y_train, y_test)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier
```

```python
# Apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
# Random Forest Hyperparameter Tuning with RandomizedSearchCV
rf_model = RandomForestClassifier(random_state=42)
rf_param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
```

```python
rf_search = RandomizedSearchCV(rf_model, rf_param_dist, n_iter=10, cv=3, random_state=42)
rf_search.fit(X_resampled, y_resampled)
```

```python
# Best Random Forest Model
print("Best Random Forest Parameters:", rf_search.best_params_)
best_rf_model = rf_search.best_estimator_
```

```python
# XGBoost Hyperparameter Tuning
xgb_model = XGBClassifier(random_state=42)
xgb_param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.3],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
}
```

```python
xgb_search = RandomizedSearchCV(xgb_model, xgb_param_dist, n_iter=10, cv=3, random_state=42)
xgb_search.fit(X_resampled, y_resampled)
```

```python
# Best XGBoost Model
print("Best XGBoost Parameters:", xgb_search.best_params_)
best_xgb_model = xgb_search.best_estimator_
```

```python
# Evaluate the best models on the test set
rf_pred = best_rf_model.predict(X_test)
xgb_pred = best_xgb_model.predict(X_test)
```

```python
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print("XGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
```
